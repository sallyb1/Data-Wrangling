```python
# set a function for tweet extraction
# file already created so no need to execute to continue the notebook
count = 0
fails_dict = {}
start = time.time()
with open('tweet-json.txt', 'w') as outfile:
    for tweet_id in tweet_ids:
        count += 1
        print(str(count) + ": " + str(tweet_id))
        try:
            tweet = api.get_status(tweet_id, tweet_mode='extended')
            print("Success")
            json.dump(tweet._json, outfile)
            outfile.write('\n')
        except tweepy.errors.TweepyException as e:
            print("Fail")
            fails_dict[tweet_id] = e
            pass
end = time.time()
print(end - start)
print(fails_dict)
```

```python
# create empty lists to save our choice for each row in the dataset
breed = []
confidence = []

# function that iterates through prediction columns to find the best prediction which is a breed of dog.
def breed_confidence(row):
    if row.p1_dog == True:
        breed.append(row.p1)
        confidence.append(row.p1_conf)
    elif row.p2_dog == True:
        breed.append(row.p2)
        confidence.append(row.p2_conf)
    elif row.p3_dog == True:
        breed.append(row.p3)
        confidence.append(row.p3_conf)
    else:
        breed.append('Unpredicted')
        confidence.append(0)
        
# call function using pandas apply by columns
image_clean.apply(breed_confidence, axis=1)

# add lists created to master dataframe
image_clean['breed'] = breed
image_clean['confidence'] = confidence   
```

### Tidiness issues
#### Define
- Convert the ["doggo", "flooter", "pupper", "puppo"] columns into one "stage" column, then drop the four columns


#### Code

```python
master['stage']=None
master['stage'] = master.doggo + master.floof + master.pupper + master.puppo
```

```python
master['stage'].value_counts()
```

```python
# create function to check dog stages
def check_stages(master):
    if master['stage'].count('None') == 2:
        return 'Multiple' #this means it has more than one dog stage
    else:
        if master['stage'].count('doggo') == 1:
            return 'Doggo'
        elif master['stage'].count('floofer') == 1:
            return 'Floofer'
        elif master['stage'].count('pupper') == 1:
            return 'Pupper'
        elif master['stage'].count('puppo') == 1:
            return 'Puppo'
        else:
            return 'None'

master['stage'] = master.apply(check_stages, axis=1)
```

### Quality Issue
#### Define 
- Remove hyperlinks

#### Code

```python
#define function and apply to archive_clean table
def htmlink(x):
        http_pos = x.find("http")
        # If no link, retain row
        if http_pos == -1:
            x = x
        else:
            # Remove space before link to end
            x = x[:http_pos - 1]
        return x
    
master.text = master.text.apply(htmlink)
```

```python
from wordcloud import WordCloud, STOPWORDS
from PIL import Image
import urllib
```

```python
tweets = np.array(master.text)
my_list = []
for tweet in tweets:
    my_list.append(tweet.replace("\n",""))
```

```python
mask = np.array(Image.open(requests.get('http://www.clker.com/cliparts/1/4/2/8/11954403871637743121zeimusu_dog_face_4.svg.med.png',stream=True).raw))
text = my_list
```

```python
def generate_wordcloud(text, mask):
    word_cloud = WordCloud(width = 512, height = 512, background_color='white', stopwords=STOPWORDS, mask=mask).generate(str(text))
    plt.figure(figsize=(10,8),facecolor = 'white', edgecolor='blue')
    plt.imshow(word_cloud)
    plt.axis('off')
    plt.tight_layout(pad=0)
    plt.show()
```

```python
generate_wordcloud(text,mask)
```
